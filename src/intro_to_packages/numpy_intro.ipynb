## Numpy

**NumPy (Numerical Python)** is a powerful library for numerical and scientific computing in Python. It provides support for multi-dimensional arrays and matrices, along with performing mathematical operations efficiently.

In Python we have lists that serve the same purpose as arrays, but they are slow to process. NumPy aims to provide an array object that is up to 50x faster than traditional Python lists. While Python lists are flexible, they are slow for numerical computations. NumPy arrays are faster, Memory Efficient, Convenient, support vectorization, etc.

The array object in NumPy is called ndarray. Arrays are very frequently used in the AI field, where speed and resources are very important.

```python
# Import library
import numpy as np
```

### Numpy arrays


* 0D Array (Scalar) - this holds a single value

```python
arr = np.array(3)

print(arr)
print(type(arr))
print(arr.shape)
```

* 1D Array(Vector) - 
This is a simple, linear list of elements, all of the same data type, stored sequentially in a single row.

```python
arr1 = np.array([1, 2, 3, 4, 5])

print(arr1)
print(arr1.shape) # From the output ther is just 5 columns
```

* 2D Array(Matrix) - Stores data in a table like format

```python
arr2 = np.array([[1, 2], [3, 4], [5, 6]])

print(arr2)
print(arr2.shape)
```

* 3D Array(Tensor) - Similar to stacking multiple tables on top of each other

```python
arr3 = np.array([
    [[1, 2, 3, 4], [5, 6, 7, 8]],
    [[0, 1, 2, 3], [4, 5, 6, 7]],
    [[9, 8, 7, 6], [2, 3, 5, 6]]
])

print(arr3)

print(arr3.shape)
```

#### Accessing Elements in Arrays

#### Modifying Elements

```python
# Let's do some arithmetic operations on two arrays and see the output
arr_mod1 = np.array([2, 4, 6])
arr_mod2 = np.array([8, 10, 12])

# Let's add
print(f"add_result: {arr_mod1 + arr_mod2}")
print()

# Let's multiply
print(f"multiply_result: {arr_mod1 * arr_mod2}")
print()

# Let's divide
print(f"divide_result: {arr_mod1 / arr_mod2}")
```

```python
# Let's modify an element in "arr_mod1"
arr_mod1[2] = 8
print(f"modified_arr: {arr_mod1}")
```

#### Reshaping Arrays

```python
# Define an array
arr_def = np.array([1, 2, 3, 4, 5, 6,])

# Lets reshape to a 2 by 3 array using .reshape()
reshaped_arr = arr_def.reshape(2, 3)
print(reshaped_arr)
```

```python
# It is also possible to flatten multi-dimensional arrays to 1 dimension.
flattened_arr = reshaped_arr.flatten()

print(flattened_arr)
print(flattened_arr.shape)
```

#### Aggregation and Statistical Operatons

```python
# Define an array
arr_new = np.array([1, 2, 3, 4, 5])
```

```python
# Let's sum all the elements in the array
print(np.sum(arr_new))
```

```python
# Let's take the mean of the array
print(np.mean(arr_new))
```

```python
# Let's check the median number in the array
print(np.median(arr_new))
```

```python
# We can also check the standard deviation
print(np.std(arr_new))
```

#### Random stuffs


```python
main_arr = np.random.randint(1, 100, (3, 4))

print(main_arr)
```

```python
# First 2 rows
print(main_arr[:2])
```

```python
# Last 3 columns
print(main_arr[:, -3:]) # first : selects all rows, the second ':' the last 3 columns
```

```python
# Get all even numbers in the array
print(main_arr[main_arr % 2 == 0]) # the [] contains the constraint
```

```python
# Convert to 3D array (2 * 2 * 3)
new_main = main_arr.reshape(2, 2, 3)

print(new_main)
```

```python
# Flatten to 1D
new_main1 = new_main.flatten()

print(new_main1)
```

```python
# Generate a 4x4 identity matrix
id_mat = np.eye(4)

print(id_mat)
```

```python
# Replace 1 with random numbers
np.fill_diagonal(id_mat, np.arange(10, 50, 10))

print(id_mat)
```

```python
# Random 3*4 and 4*3 matrices
arr34 = np.random.randint(1, 10, (3, 4))
arr43 = np.random.randint(1, 10, (4, 3))

print(arr34)
print(arr43)
```

```python
# Matrix multiplication
print(arr34 @ arr43)
```

```python
# Function receives a 2D array as input and then returns a new array
# where all even numbers are squared and all odd numbers are replaced with -1
def array_mod(arr):
    return np.where(arr % 2 == 0, arr ** 2, -1)
```

```python
trial_matrix = np.random.randint(1, 10, (3, 3))

print(trial_matrix)
```

```python
mod_matrix = array_mod(trial_matrix)
print(mod_matrix)
```
